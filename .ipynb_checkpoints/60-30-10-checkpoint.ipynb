# Assessing images' compliance to the 60-30-10 rule

## Problem definition

The 60-30-10 rule is a design principle used to help create balenced, cohesive colour schemes by dividing a space into 60% dominant colour, 30% secondary colour, and 10% accent colour. 

An example of it's use could be in interior design where 60% of the room (walls, flooring, large furniture pieces) is made of of the dominant colour, 30% of the room is made up of the secondary colour (curtains, rugs, accent walls, small furniture items), and 10% of the room is made up of the accent colour (throw pillows, art pieces, lampshades, decorations). 

The purpose of this project is to use image clustering algorithms to assess and quantify the level to which different images comply with this rule. 

## Libraries and Setup

Importing required packages

```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage import color
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
```

## Functions for Retrieving Images

```python
def get_image_data(source, max_pixels=10000000):
    if source.startswith("http://") or source.startswith("https://"):
        response = requests.get(source)
        img = Image.open(BytesIO(response.content)).convert("RGB")
    else:
        img = Image.open(source).convert("RGB")
    
    imgwidth, imgheight = img.size
    pixelcount = imgwidth * imgheight
    
    if pixelcount > max_pixels:
        scale = 1/np.sqrt(pixelcount / max_pixels)
        imgwidth = max(int(np.floor(imgwidth*scale)), 1)
        imgheight = max(int(np.floor(imgheight*scale)), 1)
        img = img.resize((imgwidth, imgheight), Image.LANCZOS)

    # imgwidth = img.size[0]
    # imgheight = img.size[1]
    # pixelcount = imgwidth * imgheight
    
    return img
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(img)

k = 3
kmeans = KMeans(n_clusters=k, n_init=5)

# Fit/predict on hue only
hue_values = pixels_hsv[:, 0]
hue_transformed = np.stack([
    np.cos(2 * np.pi * hue_values),
    np.sin(2 * np.pi * hue_values)
], axis=1)
labels = kmeans.fit_predict(hue_transformed)

sil = silhouette_score(hue_transformed, labels)
db = davies_bouldin_score(hue_transformed, labels)
ch = calinski_harabasz_score(hue_transformed, labels)

# Get hue centroids
hue_centers = np.arctan2(
    kmeans.cluster_centers_[:, 1],
    kmeans.cluster_centers_[:, 0]
) / (2 * np.pi) % 1.0  # normalise back to [0, 1]

# Replace each pixel's hue with its cluster's hue centroid, keep original S and V
pixels_hsv_new = pixels_hsv.copy()
pixels_hsv_new[:, 0] = hue_centers[labels]

# Convert back to RGB for image display
pixels_kmeans = np.uint8(color.hsv2rgb(pixels_hsv_new) * 255).reshape(img.size[1], img.size[0], 3)
pixels_hsv_new[:, 1] = 0.8  # saturation
pixels_hsv_new[:, 2] = 0.8  # value
pixels_kmeans_hue = np.uint8(color.hsv2rgb(pixels_hsv_new) * 255).reshape(img.size[1], img.size[0], 3)

plt.subplot(1, 3, 2)
plt.imshow(pixels_kmeans)


plt.subplot(1, 3, 3)
plt.imshow(pixels_kmeans_hue)

# For bar chart: build centers with centroid hue, S=1, V=1 for pure colour swatches
bar_hsv_centers = np.stack([hue_centers, np.full(k, 0.8), np.full(k, 0.8)], axis=1)  # shape (k, 3)
centers = np.uint8(color.hsv2rgb(bar_hsv_centers) * 255)

counts = np.bincount(labels)
idx = np.argsort(counts)[::-1]

percentages = counts[idx] / counts.sum() * 100

bar_labels = [f"{(360 * hue_centers[idx[i]]):.1f}" for i in range(k)]
plt.figure(figsize=(10, 6))
plt.bar(bar_labels, percentages, color=centers[idx] / 255)
plt.xticks(rotation=90)
plt.xlabel("Color (R,G,B)")
plt.ylabel("Percentage (%)")
plt.title(f"{k} cluster centers (hue-only clustering)")
plt.ylim(0, 100)
plt.show()

sil, db, ch
```

```python
def enforce_ratio(labels, hue_transformed, k, ratio=(0.60, 0.30, 0.10)):
    n_pixels = len(labels)
    labels = labels.copy()
    
    # Target counts
    targets = np.array([int(r * n_pixels) for r in ratio])
    # Fix rounding so targets sum to n_pixels
    targets[-1] = n_pixels - targets[:-1].sum()
    
    # Assign cluster indices to roles (0=dominant, 1=secondary, 2=accent)
    # based on current cluster sizes largest->smallest
    counts = np.bincount(labels, minlength=k)
    size_order = np.argsort(counts)[::-1]  # cluster ids sorted by size
    
    role_to_cluster = size_order  # role 0 (dominant) -> largest cluster id
    cluster_to_role = np.zeros(k, dtype=int)
    cluster_to_role[size_order[0]] = 0  # largest -> dominant
    cluster_to_role[size_order[1]] = 1  # second  -> secondary  
    cluster_to_role[size_order[2]] = 2  # smallest -> accent
    
    # Remap labels to roles so label 0=dominant, 1=secondary, 2=accent
    labels = cluster_to_role[labels]
    # Remap hue_transformed cluster centers to match role ordering
    centers_reordered = hue_transformed  # distances computed directly below

    def get_counts():
        return np.bincount(labels, minlength=3)

    def distance_to_center(pixel_indices, role):
        cluster_id = role_to_cluster[role]
        # Centroid in transformed space is mean of all points currently in that role
        center = hue_transformed[labels == role].mean(axis=0)
        return np.linalg.norm(hue_transformed[pixel_indices] - center, axis=1)

    def reassign(from_role, to_roles):
        """
        Reassign pixels from from_role to to_roles until each to_role hits its target.
        to_roles is an ordered list of roles to fill.
        """
        for to_role in to_roles:
            current_counts = get_counts()
            deficit = targets[to_role] - current_counts[to_role]
            if deficit <= 0:
                continue
            
            # Indices of pixels currently in from_role
            candidate_indices = np.where(labels == from_role)[0]
            
            if len(candidate_indices) == 0:
                break

            # Compute distance of each candidate to the to_role center
            dists = distance_to_center(candidate_indices, to_role)
            
            # Sort by closest first
            sorted_order = np.argsort(dists)
            to_reassign = candidate_indices[sorted_order[:deficit]]
            labels[to_reassign] = to_role

    counts = get_counts()
    over = [i for i in range(3) if counts[i] > targets[i]]
    under = [i for i in range(3) if counts[i] < targets[i]]

    if len(over) == 2:
        # Two clusters oversized - reassign from both to the undersized one
        # Fill the undersized cluster from whichever oversized cluster's pixels are closest
        under_role = under[0]
        under_idx = np.where((labels == over[0]) | (labels == over[1]))[0]
        
        # For each candidate pixel compute distance to the undersized center
        center_under = hue_transformed[labels == under_role].mean(axis=0)
        dists = np.linalg.norm(hue_transformed[under_idx] - center_under, axis=1)
        sorted_order = np.argsort(dists)
        
        deficit = targets[under_role] - get_counts()[under_role]
        to_reassign = under_idx[sorted_order[:deficit]]
        labels[to_reassign] = under_role
        
        # Now one oversized cluster remains - reassign its excess to the other under cluster
        counts = get_counts()
        over = [i for i in range(3) if counts[i] > targets[i]]
        under = [i for i in range(3) if counts[i] < targets[i]]
        if over and under:
            reassign(over[0], under)

    elif len(over) == 1:
        # One cluster oversized - compare each of its pixels to both undersized centers
        over_role = over[0]
        candidate_indices = np.where(labels == over_role)[0]
        
        # For each candidate compute distance to each undersized cluster center
        # and track which cluster it's closest to
        dist_matrix = np.stack([
            distance_to_center(candidate_indices, u) for u in under
        ], axis=1)  # shape (n_candidates, n_under)
        
        # Closest undersized cluster for each pixel
        closest_under = np.argmin(dist_matrix, axis=1)  # index into `under`
        min_dists = dist_matrix[np.arange(len(candidate_indices)), closest_under]
        
        # Sort all candidates by their distance to their closest undersized center
        sorted_order = np.argsort(min_dists)
        
        original_under = under.copy()  # cache before loop

        for pos in sorted_order:
            counts = get_counts()
            under = [i for i in range(3) if counts[i] < targets[i]]
            if not under:
                break
            pixel_idx = candidate_indices[pos]
            # Use original_under to index closest_under safely
            preferred = original_under[closest_under[pos]]
            # Assign to preferred if still undersized, otherwise any remaining undersized cluster
            if preferred in under:
                labels[pixel_idx] = preferred
            else:
                labels[pixel_idx] = under[0]

    return labels, role_to_cluster, targets
```

```python
labels_enforced, role_to_cluster, targets = enforce_ratio(labels, hue_transformed, k=3)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(img)

sil = silhouette_score(hue_transformed, labels_enforced)
db = davies_bouldin_score(hue_transformed, labels_enforced)
ch = calinski_harabasz_score(hue_transformed, labels_enforced)

# Get hue centroids
hue_centers = np.arctan2(
    kmeans.cluster_centers_[:, 1],
    kmeans.cluster_centers_[:, 0]
) / (2 * np.pi) % 1.0  # normalise back to [0, 1]

# Replace each pixel's hue with its cluster's hue centroid, keep original S and V
pixels_hsv_new = pixels_hsv.copy()
pixels_hsv_new[:, 0] = hue_centers[labels_enforced]

# Convert back to RGB for image display
pixels_kmeans = np.uint8(color.hsv2rgb(pixels_hsv_new) * 255).reshape(img.size[1], img.size[0], 3)
pixels_hsv_new[:, 1] = 0.8  # saturation
pixels_hsv_new[:, 2] = 0.8  # value
pixels_kmeans_hue = np.uint8(color.hsv2rgb(pixels_hsv_new) * 255).reshape(img.size[1], img.size[0], 3)

plt.subplot(1, 3, 2)
plt.imshow(pixels_kmeans)


plt.subplot(1, 3, 3)
plt.imshow(pixels_kmeans_hue)

# For bar chart: build centers with centroid hue, S=1, V=1 for pure colour swatches
bar_hsv_centers = np.stack([hue_centers, np.full(k, 0.8), np.full(k, 0.8)], axis=1)  # shape (k, 3)
centers = np.uint8(color.hsv2rgb(bar_hsv_centers) * 255)

counts = np.bincount(labels_enforced)
idx = np.argsort(counts)[::-1]

percentages = counts[idx] / counts.sum() * 100

bar_labels = [f"{(360 * hue_centers[idx[i]]):.1f}" for i in range(k)]
plt.figure(figsize=(10, 6))
plt.bar(bar_labels, percentages, color=centers[idx] / 255)
plt.xticks(rotation=90)
plt.xlabel("Color (R,G,B)")
plt.ylabel("Percentage (%)")
plt.title(f"{k} cluster centers (hue-only clustering)")
plt.ylim(0, 100)
plt.show()
```
